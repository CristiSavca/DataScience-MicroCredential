```python
# Using pandasql

query = "SELECT * FROM df"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE Gender == 'Male' AND Age == 30"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE Gender == 'Male' AND EstimatedSalary >= 150000"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE EstimatedSalary >= 130000"
ps.sqldf(query)
```
